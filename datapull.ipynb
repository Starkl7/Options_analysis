```python
period = "1mo"
interval = "1h"
aapl = yf.Ticker("AAPL")
aapl_stock = aapl.history(period=period, interval=interval)
```

```python
stock_data = {
    "aapl": { 
        "stock": aapl_stock,
        "period": period,
        "interval": interval
    }
}

options_data = {
    "aapl": {
        "option_chain": opt_chain,
        "expiration_date": aapl.options[0]
    }
}

def get_data(ticker):
    return {
        "stock": stock_data.get(ticker),
        "options": options_data.get(ticker)
    }
```
